### 01) WAP to demonstrate the use of Pie chart.

```python
import matplotlib.pyplot as plt
%matplotlib inline
values = [305,201,805,75,436]
l = ["Food","Travel","Accomodation","Misc","Shopping"];
c = ['red','green','blue','pink','yellow']
e = [0,0.2,0,0,0]
plt.pie(values,colors=c,labels=l,shadow='true',autopct="%1.1f%%")
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt
%matplotlib inline
x = [1,2,3,4,5]
y = [5.9,6.2,3.2,8.9,9.7]
l = ['Grade-A','Grade-B','Grade-C','Grade-D','Grade-E']
c = ['b','r','g','y','m']
plt.title("Grade Sheet")
plt.bar(x,y,color=c,label=l)
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
cpis = np.random.randint(0,10,100)
plt.hist(cpis,bins=10,
histtype='stepfilled',align='mid',label='CPI List')
plt.legend()
plt.show();
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
x = ["A", "B", "C", "D"]
y = [1, 2, 3, 4]
plt.barh(x, y)
for index, value in enumerate(y):
    plt.text(value, index,str(value))
plt.show()
```
